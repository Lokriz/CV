```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
gas_data = pd.read_csv('gas_prices.csv')
gas_data
```

```python
plt.figure(figsize = [10,7])
for country in gas_data.iloc[:,1:]:
    plt.plot(gas_data.Year, gas_data[country], '.-')

plt.title('Цены на газ', fontdict = {'fontweight': 'bold'})
plt.xlabel('Год', fontsize = 12)
plt.ylabel('Цена, $')
plt.grid()
plt.xticks(gas_data.Year[::2])
plt.yticks(list(range(0,8)))
plt.legend(gas_data.columns[1:])
plt.show()
```

### Анализ данных 

```python
fifa = pd.read_csv('fifa_data.csv')
print(fifa.columns)
fifa = fifa.iloc[:,list(range(1,4)) + [5] + list(range(7,10)) + [14, 27, 11]]
fifa
```

```python
bins = list(range(40, 101, 10))

plt.hist(fifa.Overall, bins = bins)
plt.grid(color='white', lw = 1, axis = 'x')

plt.xticks(bins)
plt.title('Гистограмма навыков')
plt.xlabel('Оценка навыка')
plt.ylabel('Кол-во игроков')
plt.show()
```

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]
non = fifa.count()[0] - left - right
```

```python
plt.pie([left, right, non], labels = list(fifa.groupby('Preferred Foot').count().iloc[:,0].index) + ['NM'], autopct = '%.2f %%')
plt.title('Круговая диаграмма преобладающих ног')
plt.show()

list(fifa.groupby('Preferred Foot').count().iloc[:,0].index)
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight]

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa.loc[fifa.Weight >= 200].count()[0]

lbl = ['До 125 фунтов', '125-150 фунтов', '150-175 фунтов', '175-200 фунтов', 'Более 200 фунтов']
plt.figure(figsize = (10,10))
plt.title('Weight cathegories of FIFA players', fontsize = 22)
plt.pie([light, light_medium, medium, medium_heavy, heavy], labels = lbl, autopct = '%.2f %%', explode = [.4, 0.4, 0, 0, 0.4], pctdistance = 0.85)


plt.show()
```

```python
plt.figure(figsize=(8, 8))

barce = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
juve = fifa.loc[fifa.Club == 'Juventus']['Overall']
psg = fifa.loc[fifa.Club == 'Paris Saint-Germain']['Overall']

boxes = plt.boxplot([barce, juve, psg], labels = ['ФК Барселона', 'Ювентус', 'Пари Сен-Жермен'])
for box in boxes['boxes']:
    box. set(color = '#4286f2', linewidth = 2)
    
plt.ylabel('Рэйтинг')
plt.title('Сравнение команд', fontsize = 22, fontdict = {'fontweight': 'bold'})
    

plt.show()
```
